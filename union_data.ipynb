```python
import pandas as pd
import random
from faker import Faker
from datetime import datetime, timedelta

# Initialize Faker for generating fake data
fake = Faker()

# Create empty dataframes for each entity
customers = pd.DataFrame(columns=['CustomerID', 'Name', 'Email', 'ContactNumber'])
addresses = pd.DataFrame(columns=['AddressID', 'CustomerID', 'StreetAddress', 'City', 'State', 'PostalCode'])
products = pd.DataFrame(columns=['ProductID', 'Name', 'Description', 'Category'])
variants = pd.DataFrame(columns=['VariantID', 'ProductID', 'Color', 'Size', 'Weight'])
prices = pd.DataFrame(columns=['PriceID', 'ProductID', 'VariantID', 'Price', 'StartDate', 'EndDate'])
orders = pd.DataFrame(columns=['OrderID', 'CustomerID', 'OrderDate', 'ShippingAddressID'])
order_details = pd.DataFrame(columns=['OrderDetailID', 'OrderID', 'ProductID', 'VariantID', 'Quantity', 'UnitPrice'])

# Generate sample data
for i in range(10):  # Create 10 customers
    customers = customers.append({
        'CustomerID': i + 1,
        'Name': fake.name(),
        'Email': fake.email(),
        'ContactNumber': fake.phone_number(),
    }, ignore_index=True)

    addresses = addresses.append({
        'AddressID': i + 1,
        'CustomerID': i + 1,
        'StreetAddress': fake.street_address(),
        'City': fake.city(),
        'State': fake.state(),
        'PostalCode': fake.zipcode(),
    }, ignore_index=True)

for i in range(10):  # Create 10 products
    products = products.append({
        'ProductID': i + 1,
        'Name': fake.word(),
        'Description': fake.sentence(),
        'Category': fake.word(),
    }, ignore_index=True)

    variants = variants.append({
        'VariantID': i + 1,
        'ProductID': i + 1,
        'Color': fake.color_name(),
        'Size': fake.random_element(elements=('Small', 'Medium', 'Large')),
        'Weight': random.uniform(0.1, 10.0),  # Random weight between 0.1 and 10.0
    }, ignore_index=True)

for i in range(10):  # Create 10 prices (historical data)
    start_date = datetime.now() - timedelta(days=random.randint(1, 730))  # Within the last 2 years
    end_date = start_date + timedelta(days=random.randint(1, 30))  # Up to 30 days from start_date

    prices = prices.append({
        'PriceID': i + 1,
        'ProductID': random.randint(1, 10),  # Randomly link to a product
        'VariantID': random.randint(1, 10),  # Random
        'Price': random.uniform(10.0, 100.0),  # Random price between 10.0 and 100.0
        'StartDate': start_date,
        'EndDate': end_date,
    }, ignore_index=True)

for i in range(10):  # Create 10 orders
    order_date = datetime.now() - timedelta(days=random.randint(1, 730))  # Within the last 2 years
    orders = orders.append({
        'OrderID': i + 1,
        'CustomerID': random.randint(1, 10),  # Randomly link to a customer
        'OrderDate': order_date,
        'ShippingAddressID': random.randint(1, 10),  # Randomly link to an address
    }, ignore_index=True)

for i in range(20):  # Create 20 order details (products within orders)
    order_details = order_details.append({
        'OrderDetailID': i + 1,
        'OrderID': random.randint(1, 10),  # Randomly link to an order
        'ProductID': random.randint(1, 10),  # Randomly link to a product
        'VariantID': random.randint(1, 10),  # Randomly link to a variant
        'Quantity': random.randint(1, 5),  # Random quantity between 1 and 5
        'UnitPrice': random.uniform(10.0, 100.0),  # Random price between 10.0 and 100.0
    }, ignore_index=True)

# Print a few rows from each DataFrame to check
print(customers.head())
print(addresses.head())
print(products.head())
print(variants.head())
print(prices.head())
print(orders.head())
print(order_details.head())

# Save DataFrames to CSV files
customers.to_csv('customers.csv', index=False)
addresses.to_csv('addresses.csv', index=False)
products.to_csv('products.csv', index=False)
variants.to_csv('variants.csv', index=False)
prices.to_csv('prices.csv', index=False)
orders.to_csv('orders.csv', index=False)
order_details.to_csv('order_details.csv', index=False)
```
