```python
import pandas as pd

# Read the Excel file into a pandas DataFrame
df = pd.read_excel('TimeSeries.xlsx')  # Replace 'your_file.xlsx' with the path to your Excel file

# Sort the DataFrame by bot name and start time
df.sort_values(by=['Name', 'Start'], inplace=True)

# Initialize variables to track the current period
current_bot = ''
period_start = ''
period_end = ''
activities = []

# Lists to store transformed data
periods = []
aggregated_activities = []

# Iterate through rows in the DataFrame
for index, row in df.iterrows():
    if row['Name'] != current_bot:
        # If the bot name changes, start a new period
        if current_bot:
            periods.append((current_bot, period_start, period_end))
            aggregated_activities.append(', '.join(activities))
        current_bot = row['Name']
        period_start = row['Start']
        period_end = row['End']
        activities = [row['Activity']]
    else:
        # If the bot name is the same, extend the current period
        period_end = row['End']
        activities.append(row['Activity'])

# Append the last period to the lists
periods.append((current_bot, period_start, period_end))
aggregated_activities.append(', '.join(activities))

# Create a new DataFrame for the transformed data
transformed_df = pd.DataFrame({'Name': [p[0] for p in periods],
                               'Period Start': [p[1] for p in periods],
                               'Period End': [p[2] for p in periods],
                               'Activities': aggregated_activities})

# Save the transformed data to a new Excel file or use it as needed
transformed_df.to_excel('transformed_data.xlsx', index=False)  # Replace with your desired output file path

```
